# Contexte
    
Amateur de tennis, je suis les saisons attentivement et j'aime également parier dessus. S'il existe certaines statistiques pouvant aider à la décision, on regardera dans un premier temps, le classement des joueurs, leurs ratios de victoires, leurs historiques, etc... On peut évidemment trouver dans les ratios de victoires, des ratios par surface, par niveau de tournoi, par classement (contre des top 10, top 20, etc...). Ce dernier point me semble important, puisque le classement permet de comparer les joueurs entre eux et d'indiquer si tel joueur est meilleur qu'un autre, sur le papier.

Le classement est évolutif et change toutes les semaines en ajoutant les points fraichement gagnés lors de la semaine passée et en enlevant les points gagnés l'année précédente lors de la semaine équivalente. Il existe 4 types de tournois avec un barème spécifique de points de manière exponentielle en fonction du palier atteint dans le tournois. Les 250 (250 pts max en cas de victoire finale), les 500 (500 pts), les Masters 1000 (1000pts) et les Grands Chelems (2000 pts). Dans la même semaine, il peut se dérouler plusieurs tournois 250 et/ou 500, alors que les M1000 et les GC occupent 1 ou 2 semaines à eux seuls. Chacun est ± libre de configurer son calendrier pour participer aux tournois qu'il souhaite, et focément plus le tournoi est important, plus il y aura de bons joueurs.
   
Une saison c'est long et pour un joueur, il existe différents facteurs pouvant altérer son classement. Du coté négatif, il y a évidemment les blessures pouvant coûter au joueur, une absence de compétition et donc aucun gain de points, alors que les autres joueurs continuent d'en engranger (si par malheur, la blessure est grave et tient le joueur éloignés du circuit pendant plus un long moment, le joueur sera forcément très en dessous de son classement potentiel qu'il aurait pu obtenir s'il avait joué tout au long de la saison, voire non classé si son absence est superieure à 1 an). Du coté positif, le facteur chance, qui par un tableau dégagé grâce à plusieurs contre-perfs ou forfait des meilleurs joueuses, ou tout simplement par forfait de son adversaire direct (en cours de match ou juste avant), ou par un tableau de départ relativement faible par rapport à l'importance du tournoi. Ce facteur chance peut donc apporter un gain de points supérieur à celui qu'on aurait du avoir "normalement" et si c'est un gros tournoi, et bien le gain "chanceux" de points peut devenir extrêmement conséquent et propulser un joueur très fortement dans le haut du classement. Le facteur chance peut aussi être un facteur surperformance, où un joueur va gagner sur un tournoi important 1 ou 2 matchs clés en offrant une prestation extraordinaire.

Le classement n'est pas seulement indicatif, il peut aussi offrir, pour les meilleurs joueurs, une place de tête de série permettant d'avoir, selon le tournoi, un matchs en moins et dans tous les cas un meilleur tirage (par ex: les 2 joueurs les mieux classés d'un tournoi ne peuvent s'affronter qu'en finale).
    
Le tennis se joue sur 3 types de surface différents, avec des joueurs ± à l'aise selon le type de surface et le classement prend en compte tous les résultats quelque soit la surface. En prenant en compte tous ces paramètres, le classement ne sera "jamais juste" et comportera des anomalies.
    
C'est ainsi que je me suis interrogé sur la création d'une Tier List (les détails seront abordés plus tard) permettant de regrouper les joueurs de niveau ± équivalent, ce qui simplifirai les rapports de forces entre 2 joueurs et masquerait les défauts de ce classement.     
    
Remarque : Je n'ai que parlé du circuit principal composé de 250, 500, M1000 et GC, mais il existe d'autres compétitions permettant d'obtenir de points comme L'ATP/WTA FINALS (avec les 8 meilleurs joueurs/ses) ou encore la United Cup et les circuits secondaires et tertiaires avec respectivement les tournois challengers et les tournois futurs (avec un barème de points inférieur)

### Tier List
    
La création de la Tier List (TL) est basée par mes connaissances et mon ressenti personnel, elle est donc totalement subjective. Je ne me suis pas fixé de critères particuliers pour la composition de ces TL, mais principalement je tiens en compte la régularité contre les joueurs de TL inférieures et le rapport de force contre les joueurs appartenant à la même TL et aux TL voisines.
    
Les critères seront, dans tous les cas, propres à chaque TL et l'objectif est de minimiser la variation de niveau des joueurs appartenant à une même TL.
    
Forcément cette TL est loin d'être optimale, notamment à cause de différents facteurs comme : le classement, mes souvenirs lointains ou récents, mon affection personnelle ou encore le nom/statut d'un joueur.

Finalement, il y a encore des variables qui ne me seront d'aucune utilité, je vais donc les enlever afin d'y voir plus clair, puis j'utiliserai la méthode groupby pour accomplir l'objectif

```python
tennis_m.drop(columns={'winner_name','loser_name','tourney_name','tourney_date','winner_ioc','winner_age','loser_ioc',
                       'loser_age','round','score','winner_rank','loser_rank','match_num'},inplace=True)
tennis_m.head()
```

```python
# Création d'un DF groupé "Winner" 
tennis_w = tennis_m.drop(columns={'loser_id'}).copy()
tennis_w['win'] = 1
tennis_w = tennis_w.groupby(by=['winner_id','surface','tourney_level','loser_tl'],as_index=False).sum()

# Création DF groupé "Loser"
tennis_l = tennis_m.drop(columns={'winner_id'}).copy()
tennis_l['lose'] = 1
tennis_l = tennis_l.groupby(by=['loser_id','surface','tourney_level','winner_tl'],as_index=False).sum()

# Jointure des DF "Winner" et "Loser"
tennis_w.rename(columns={'winner_id':'player_id','loser_tl':'tier_list'},inplace=True)
tennis_l.rename(columns={'loser_id':'player_id','winner_tl':'tier_list'},inplace=True)
tennis_wl = pd.merge(tennis_w,tennis_l,how='outer',on=['player_id','tourney_level','surface','tier_list'])

# Valeurs nulles remplacées par 0  
tennis_wl.fillna(0,inplace=True)

# Transformation de décimal à entier
tennis_wl['win'] = np.int64(tennis_wl['win'])
tennis_wl['lose'] = np.int64(tennis_wl['lose'])

tennis_wl.head()
```
